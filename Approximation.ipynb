```python
import numpy as np
from Graphe_OO.Graph import Graph
from Graphe_OO.Algorithmes_Approches import AlgoMatrice
from Graphe_OO.Heuristique_BAB_OO import *
import pandas as pd
import seaborn as sns
from stats import *
from Graphe_Adjacence.Graph2 import Graph_A
from Graphe_Adjacence.Algorithmes_Approches2 import AlgoAdjacence
from Graphe_Adjacence.Heuristique_BAB import *
```

## Matrice

```python
rglouton = approximation(AlgoMatrice.algo_glouton,parcours_profondeur_lite_ameliore_1_OO,Graph,30,0.5)
rcouplage = approximation(AlgoMatrice.algo_couplage,parcours_profondeur_lite_ameliore_1_OO,Graph,30,0.5)
```

```python

r=pd.DataFrame()
r['n']=rglouton['n']
r['glouton']=rglouton["approximation"]
r['couplage']=rcouplage['approximation']

rcat = r.melt(id_vars=['n'],value_vars=['glouton','couplage'],
              var_name='Algo', value_name='approximation')

gm = sns.lineplot(x=rcat["n"],y=rcat['approximation'],
             hue=rcat.Algo,palette=sns.color_palette("pastel", 2))

gm.get_figure().savefig("img/approximation_matrice.png")
```
