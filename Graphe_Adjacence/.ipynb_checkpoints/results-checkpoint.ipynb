# Resultats statistiques

```python
import numpy as np
from Graph2 import Graph
import Algorithmes_Approches2 as aa
import pandas as pd
import seaborn as sns
```

## Algo Glouton

```python
g = Graph(5,0.5)
print(g)

print(aa.algo_glouton(g))
```

```python
results_glouton1 = aa.timeN(aa.algo_glouton,1,100,0.5,"log(f(n))")
results_glouton2 = aa.timeN(aa.algo_glouton,1,100,0.5,"log(f(n))")

#là c'est caca
```

```python
rg=pd.DataFrame()
rg['n']=results_glouton1['n']
rg['1']=results_glouton1["log(f(n))"]
rg['2']=results_glouton2["log(f(n))"]

rgcat = rg.melt(id_vars=['n'],value_vars=['1','2'],
              var_name='test', value_name="log(f(n))")

sns.lineplot(x=rgcat["n"],y=rgcat["log(f(n))"],
             palette=sns.color_palette("pastel", 2))
```

## Algo Couplage

```python
results_couplage1 = aa.timeN(aa.algo_couplage,1,100,0.5,"log(f(n))")
results_couplage2 = aa.timeN(aa.algo_couplage,1,100,0.5,"log(f(n))")
```

```python
rc=pd.DataFrame()
rc['n']=results_couplage1['n']
rc['1']=results_couplage1["log(f(n))"]
rc['2']=results_couplage2["log(f(n))"]

rccat = rc.melt(id_vars=['n'],value_vars=['1','2'],
              var_name='test', value_name="log(f(n))")

sns.lineplot(x=rccat["n"],y=rccat["log(f(n))"],
             palette=sns.color_palette("pastel", 2))
```
