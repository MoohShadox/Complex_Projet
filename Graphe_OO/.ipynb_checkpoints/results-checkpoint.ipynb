# Resultats statistiques

```python
import numpy as np
from Graph import Graph
import Algorithmes_Approches as aa
import pandas as pd
import seaborn as sns
```

## Génération aléatoire de graph

```python
results25 = aa.timeNGeneration(1,1000,0.25)
results75 = aa.timeNGeneration(1,1000,0.75)
```

```python

r=pd.DataFrame()
r['n']=results25['n']
r['0.25']=results25['f(n)']
r['0.75']=results75['f(n)']

rcat = r.melt(id_vars=['n'],value_vars=['0.25','0.75'],
              var_name='density', value_name='f(n)')

sns.lineplot(x=rcat["n"],y=rcat["f(n)"],
             hue=rcat.density,palette=sns.color_palette("pastel", 2))
```

## Algo Glouton

```python
results_glouton1 = aa.timeN(aa.algo_glouton,1,100,0.5,"log(f(n))")
results_glouton2 = aa.timeN(aa.algo_glouton,1,100,0.5,"log(f(n))")
```

```python
rg=pd.DataFrame()
rg['n']=results_glouton1['n']
rg['1']=results_glouton1["log(f(n))"]
rg['2']=results_glouton2["log(f(n))"]

rgcat = rg.melt(id_vars=['n'],value_vars=['1','2'],
              var_name='test', value_name="log(f(n))")

sns.lineplot(x=rgcat["n"],y=rgcat["log(f(n))"],
             palette=sns.color_palette("pastel", 2))
```
