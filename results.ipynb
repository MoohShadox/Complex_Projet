## Génération aléatoire de graph

### Matrice

```python
results25 = AlgoMatrice.timeNGeneration(5,500,0.25,mesure="log(f(n))")
results75 = AlgoMatrice.timeNGeneration(5,500,0.75,mesure="log(f(n))")
```

```python

r=pd.DataFrame()
r['n']=results25['n']
r['0.25']=results25['log(f(n))']
r['0.75']=results75['log(f(n))']

rcat = r.melt(id_vars=['n'],value_vars=['0.25','0.75'],
              var_name='density', value_name='log(f(n))')

gm = sns.lineplot(x=rcat["n"],y=rcat["log(f(n))"],
             hue=rcat.density,palette=sns.color_palette("pastel", 2))

gm.get_figure().savefig("img/generation_matrice.png")
```

### Adjacence

```python
results25_A = AlgoAdjacence.timeNGeneration(1,200,0.25,mesure="log(f(n))")
results75_A = AlgoAdjacence.timeNGeneration(1,200,0.75,mesure="log(f(n))")
```

```python

r_A=pd.DataFrame()
r_A['n']=results25_A['n']
r_A['0.25']=results25_A['log(f(n))']
r_A['0.75']=results75_A['log(f(n))']

rcat_A = r_A.melt(id_vars=['n'],value_vars=['0.25','0.75'],
              var_name='density', value_name='log(f(n))')

ga = sns.lineplot(x=rcat_A["n"],y=rcat_A["log(f(n))"],
             hue=rcat_A.density,palette=sns.color_palette("pastel", 2))

ga.get_figure().savefig("img/generation_adjacence.png")
```
